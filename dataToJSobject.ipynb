```python
import pickle # to save/load objects
import numpy as np
import pandas as pd
```

```python
epochs = ['e1', 'e2', 'e3', 'e6', 'e8', 'e10', 'e15', 'e20', 'e30', 'e40', 'e50']
df = pd.DataFrame()
for e in epochs:
    f = open('obj/results_{}.pckl'.format(e), 'rb')
    d_tmp = pickle.load(f)
    d_tmp['Epoch'] = e
    df = df.append(d_tmp, ignore_index=True)
    f.close()
df = df.round(decimals=4)
#df.to_csv('obj/resultData.csv', index=False)
```

```python
models = list(df['Model'].unique())
resultData = {}
d = 1000
for e in epochs:
    resultData[e] = {m:{c:[int(v*d)/d for v in list(df[(df.Model==m) & (df.Epoch==e)][c])] for c in df.columns if c not in ['Epoch', 'Model']} for m in models}
```
